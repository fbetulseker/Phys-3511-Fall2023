## Problem 1

Stirling's Approximation for $ln(n!)$ is:

\begin{equation}
ln(n!) \approx nln(n) − n.
\end{equation}

Write a program to output $ln(n!)$ and its Stirling's approximation for $2 < n < 20$. Plot on the same graph the exact value and the approximate value as a function of n. Be sure to lable your curves and axes. **Hint**: the math module (and numpy.math) provides a factorial() method.

**Expected Plot (note this one has 2 < n < 100)**


```python
def exactv(num):
    return np.log(np.math.factorial(num))
def approxv(num):
    return num*(np.log(num))-num
list1 = np.arange(2,20)
exact_list=[]
approx_list=[]
for i in list1:
    exact_list.append(exactv(i))
    approx_list.append(approxv(i))
    
fig = plt.figure()
ax = fig.add_axes([0.2,0.1,0.8,0.8])
ax.plot(list1,exact_list,label="Exact")
ax.plot(list1,approx_list,label="Approx")
ax.set_xlabel("Input value")
ax.set_ylabel("Results")
ax.legend();
    
```

## Problem 5b

## Problem 6

Using *linspace* Generate 100 equally spaced points from 0 to 1. Then plot the cube of the values generated from *linespace* vs the *linspace* generated values.

**Expected plot(Problem 6)**


```python
xvalues = np.linspace(0,1,100)
yvalues = xvalues ** 3
fig = plt.figure()
axes = fig.add_axes([0.15,0.1,0.8,0.8])
axes.plot(xvalues,yvalues)
axes.set_xlabel('x')
axes.set_ylabel('$x^3$')
axes.set_title('$x^3$ vs. x');
```
